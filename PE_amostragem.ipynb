```python
data = pd.read_csv('data_v2.csv')
```

Removendo todas as linhas que tenham ao menos um campo nulo

```python
data = data.dropna()
data
```

Removendo colunas que não vamos usar

```python
data = data.drop(["cep", "cidade", "logradouro"], axis=1)
```

Obtendo os índices das linhas que contém dados inválidos (considerando as regras que estipulamos na aula) e removendo esses índices do dataset.

```python
invalid_rows_index = data[(data['area'] < 20) | (data['valor'] < 40000)].index
data = data.drop(invalid_rows_index)
data
```

Gerando um novo arquivo com os dados filtrados: data_v3.csv

```python
data.to_csv('data_v3.csv')
```

Dados para o cálculo da amostra estratificada:

```python
data['zona'].value_counts()
```

Percentuais de cada zona para validar a amostra:

```python
data['zona'].value_counts(normalize=True) * 100
```

Para uma amostra de $n=400$, os valores $n_h$ para cada classe são:

| Zona         | $n_h$ |
|--------------|-------|
| Centro-Sul   |       |
| Oeste        |       |
| Norte        |       |
| Centro-Oeste |       |
| Leste        |       |

Obtendo as amostras de cada estrato (zona) e compondo a amostra final

```python
# precisa adicionar os valores de nh aqui
estrato_1 = data[data['zona'] == 'Centro-Sul'].sample(10, random_state=42)
estrato_2 = data[data['zona'] == 'Oeste'].sample(10, random_state=42)
estrato_3 = data[data['zona'] == 'Norte'].sample(10, random_state=42)
estrato_4 = data[data['zona'] == 'Centro-Oeste'].sample(10, random_state=42)
estrato_5 = data[data['zona'] == 'Leste'].sample(10, random_state=42)
```

```python
amostra = pd.concat([estrato_1, estrato_2, estrato_3, estrato_4, estrato_5], ignore_index=True)
amostra
```

```python
amostra.to_csv('data_amostra.csv')
```
