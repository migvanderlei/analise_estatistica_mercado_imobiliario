```python
data = pd.read_csv('data_v2.csv')
```

Removendo linhas que tenham ao menos um campo nulo

```python
data = data.dropna()
data
```

Removendo colunas que não vamos usar

```python
data = data.drop(["cep", "cidade", "logradouro"], axis=1)
```

Obtendo os índices das linhas que contém dados inválidos (considerando as regras que estipulamos na aula) e removendo esses índices do dataset.

```python
invalid_rows_index = data[(data['area'] < 20) | (data['area'] > 2650)| (data['valor'] < 40000)].index
data = data.drop(invalid_rows_index)
data
```

Gerando um novo arquivo com os dados filtrados: data_v3.csv

```python
data.to_csv('data_v3.csv')
```

Dados para o cálculo da amostra estratificada:

```python
data['zona'].value_counts()
```

<!-- Para uma amostra de $n=400$, os valores $n_h$ para cada classe são:

| Zona         | $n_h$ |
|--------------|-------|
| Centro-Sul   |       |
| Oeste        |       |
| Norte        |       |
| Centro-Oeste |       |
| Leste        |       | -->
O intervalo da amostra é calculado como $\frac{N}{n} = round(i) $ (O arredondamento da divisão $\frac{População}{amostra}$)

Para uma amostra de $n=400$, o intervalo é: $\frac{3521}{400} = \lceil 8.8025 \rceil = 9$

Ordenando pela variável a estratificar (zona)

```python
data_sorted = data.sort_values(by=['zona'])
data_sorted
```

Obtendo o primeiro elemento da amostra e, em seguida, os demais índices.

(no primeiro sorteio, o `first_id` foi 3)

```python
first_id = np.random.randint(0, 9, 1)
first_id
```

```python
# range com início, fim e intervalo
ids = np.arange(3, 3508, 9)
```

Como o valor do intervalo foi arredondado, obtivemos 391 itens na amostra

```python
amostra = data.iloc[ids]
amostra
```

Percentuais de cada zona no dataset original para validar a amostra:

```python
(data.zona.value_counts(normalize=True) * 100).round()
```

Percentuais na amostra:

```python
(amostra.zona.value_counts(normalize=True) * 100).round()
```
